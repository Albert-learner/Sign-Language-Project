```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from gesture_dataset import GestureDataset
from model import Prev_Net
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
total_data_path = '/home/jabblee/Desktop/CRC_collections/CRC_update/2023_Gatherings/'

# Datasets and DataLoaders for train and test
train_dataset = GestureDataset(data_path = total_data_path)
test_dataset = GestureDataset(data_path = total_data_path, train = False)

train_loader = DataLoader(train_dataset, batch_size = 8, shuffle = True)
test_loader = DataLoader(test_dataset, batch_size = 1, shuffle = False)

# Model
model = Prev_Net().cuda()

# Loss
criterion = nn.CrossEntropyLoss()
# criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr = 0.001)

test_loss_container = []
test_accuracy_container = []


model.eval()
    
test_loss = 0
acc_point = 0

total_test_batch = len(test_loader)
validation_arr = []
validation_arr = np.zeros((8, 23))

with torch.no_grad():
    for test_data in test_loader:
        test_data['gesture_data'] = test_data['gesture_data'].cuda().float()
        test_data['class_label'] = test_data['class_label'].cuda().long()
        print(test_data['gesture_data'].shape, test_data['class_label'].shape)

        pred = model(test_data['gesture_data'])
        
        loss = criterion(pred, test_data['class_label'].long())
        
        correct_prediction = torch.argmax(pred, 1) == test_data['class_label']
        
        # Check if class index is within bounds before updating validation_arr
        class_idx = test_data['class_label'].item()
        if 0 <= class_idx < 8:
            validation_arr[class_idx][torch.argmax(pred, 1)] += 1
        
        test_loss += loss.item() / total_test_batch
        
        if correct_prediction.any():
            acc_point += 1
            
acc_ratio = acc_point / total_test_batch

test_loss_container.append(test_loss)
test_accuracy_container.append(acc_point / total_test_batch)

print('Accuracy Ratio : {}%'.format(acc_point / total_test_batch))
```
